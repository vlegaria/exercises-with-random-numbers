```python
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def random_plot(num_datos, min_val, max_val, num_decimals):
    lista_datos = []
    nombre_archivo = "datos_aleatorios.txt"
    with open(nombre_archivo, "w") as archivo:
        for i in tqdm(range(num_datos)):
            random.seed()  # Reiniciar la semilla en cada iteración
            valor = round(random.uniform(min_val, max_val), num_decimals)
            lista_datos.append(valor)
            archivo.write(f"{valor}\n")
    return lista_datos

def histograma(lista_datos, num_bins=10):
    fig, ax = plt.subplots()
    ax.hist(lista_datos, num_bins)
    ax.set_xlabel('número')
    ax.set_ylabel('Frecuencia')
    ax.set_title('Histograma de los números generados')
    plt.show()
```

```python
num_datos = 10000000
min_val = 1
max_val = 1000
num_decimals = 10
lista_valores = random_plot(num_datos, min_val, max_val, num_decimals)
histograma(lista_valores)
```

```python
histograma(lista_valores, num_bins=100)
```
